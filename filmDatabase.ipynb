# My learning reading datasets in with python using pandas

```python
import pandas as pd
```

For this project, I downloaded two dataset from IMBd, an internet database for film, and the goal is to read the table, handle NaN cases, and retrive values form one table to verify with the other.

Ex. 
- Read names table and look up an actor
- Retirve knownForTitles codes
- Read titles table and retrieve titles

```python
# access file path and read tsv file and convert it into a csv

names_tsv = 'name.basics.tsv'
names_table = pd.read_table(names_tsv, sep = '\t')
names_table.to_csv('names.csv',index=False)

titles_tsv  = 'title.basics.tsv'
titles_table = pd.read_table(titles_tsv, sep = '\t')
titles_table.to_csv('titles.csv', index = False)
```

Difference between TSV and CSV: tsv file in a text dataset seperated by tabs while csv is a text dataset separated by commas.

```python
titlesFile = 'titles.csv'
titlesTable = pd.read_csv(titlesFile, na_values=[r'\N'])
titlesTable.head(3)
```

# Write functions

- AllMovies(movies): takes in a movies array of codes and prints title and year from titles dataset.
- ActorMovies(actor): taken in an actor name as string and return and array of movies codes from names dataset.
- Search(name): takes in and actor name as string and trigers the above functions.

Note: In order to lookup values in dataset, the .loc fucntion from pandas was used.

```python
def allMovies(movies):
    for movie in movies:
        m = titlesTable.loc[titlesTable['tconst'] == movie, 'primaryTitle'].values[0]
        y = titlesTable.loc[titlesTable['tconst'] == movie, 'startYear'].values[0]
        print(m, y)
        
def actorMovies(actor):
    movies = namesTable.loc[namesTable['primaryName'] == actor, 'knownForTitles'].values[0]
    moviesArr = movies.split(',')
    return moviesArr

def search(name):
    moviesList = actorMovies(name);
    allMovies(moviesList)
```
